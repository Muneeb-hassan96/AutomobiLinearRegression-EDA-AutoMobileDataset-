# Converting numerical data to float

```python
df['normalized-losses']=df['normalized-losses'].astype(float)
df['bore']=df['bore'].astype(float)
df['stroke']=df['stroke'].astype(float)
df['horsepower']=df['horsepower'].astype(float)
df['peak-rpm']=df['peak-rpm'].astype(float)
df['price']=df['price'].astype(float)
```

I converted the specified columns to float (e.g., 'normalized-losses', 'bore', 'stroke', 
'horsepower', 'peak-rpm', 'price) to facilitate mathematical operations, handle missing 
values, and ensure compatibility, as these tasks often require numeric input.

# Handling Missing data values

```python
median_norm_losses = df['normalized-losses'].median()
df["normalized-losses"].replace(np.nan, median_norm_losses, inplace=True)
```

```python
mode_bore = df['bore'].mode()[0]
df["bore"].replace(np.nan, mode_bore, inplace=True)
```

```python
mode_stroke = df['stroke'].mode()[0]
df["stroke"].replace(np.nan, mode_stroke, inplace=True)
```

```python
df["horsepower"].replace(np.nan, 112, inplace=True)
```

```python
mean_peakRPM = df['peak-rpm'].mean()
df["peak-rpm"].replace(np.nan, mean_peakRPM, inplace=True) #5100
```

```python
df.dropna(subset=["price"], axis=0, inplace=True)

# reset index, because we droped rows
df.reset_index(drop=True, inplace=True)
```

# Finding the highly correlat coulmns

```python
import seaborn as sns
```

```python
# Selecting only numerical column
n_col = df.select_dtypes(include = ['float64','int64'])

# calculating corelation matrix
corr_matrix = n_col.corr()

#ploting correlation heatmap
plt.figure(figsize =(12,8))
sns.heatmap(corr_matrix, cmap = 'Blues', annot = True,cbar=True)
plt.title('Correlation')
plt.show()
```

```python
numeric_data = df.select_dtypes(include=['number'])
correlation_matrix = numeric_data.corr()

threshold = 0.5  

highly_correlated_cols = set()
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > threshold:
            colname = correlation_matrix.columns[i]
            highly_correlated_cols.add(colname)

print("Highly correlated columns:", highly_correlated_cols)
```

```python
target_column = 'price'

correlation_with_target = numeric_data.corr()[target_column]

threshold_positive = 0.5  # You can adjust this based on your requirements for positive correlation
threshold_negative = -0.5 

strong_positive_correlation_cols = correlation_with_target[correlation_with_target > threshold_positive].index.tolist()

strong_negative_correlation_cols = correlation_with_target[correlation_with_target < threshold_negative].index.tolist()

if target_column in strong_positive_correlation_cols:
    strong_positive_correlation_cols.remove(target_column)

if target_column in strong_negative_correlation_cols:
    strong_negative_correlation_cols.remove(target_column)

print(f"Columns with strong positive correlation with '{target_column}': {strong_positive_correlation_cols}")
print(f"Columns with strong negative correlation with '{target_column}': {strong_negative_correlation_cols}")
```

# Multiple Linear Regression

What if we want to predict car price using more than one variable?

If we want to use more variables in our model to predict car price, we can use Multiple Linear Regression. Multiple Linear Regression is very similar to Simple Linear Regression, but this method is used to explain the relationship between one continuous response (dependent) variable and two or more predictor (independent) variables.

From the previous section we know that other good predictors of price could be:

Horsepower,
Curb-weight,
Engine-size,
Highway-mpg

Let's develop a model using these variables as the predictor variables.

```python
Z = df[['horsepower', 'curb-weight', 'engine-size', 'highway-mpg']]
```

```python
lm.fit(Z, Y)
```

```python
lm.intercept_
```

```python
lm.coef_
```

As we saw above, we should get a final linear function with the structure:

$$
Yhat = a + b_1 X_1 + b_2 X_2 + b_3 X_3 + b_4 X_4
$$

What is the linear function we get in this example?

<b>Price</b> = -15678.742628061467 + 52.65851272 x <b>horsepower</b> + 4.69878948 x <b>curb-weight</b> + 81.95906216 x <b>engine-size</b> + 33.58258185 x <b>highway-mpg</b>
